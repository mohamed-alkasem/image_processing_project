```python
# YÃ¼z Ä°fadesi TanÄ±ma Sistemi - Optimize EdilmiÅŸ Versiyon
# Face Expression Recognition System - Optimized Version

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
import cv2
from google.colab import drive
import time

print("=" * 60)
print("YÃœZ Ä°FADESÄ° TANIMA SÄ°STEMÄ°")
print("Face Expression Recognition System")
print("=" * 60)

# Mount Google Drive
print("\n[1/9] Google Drive baÄŸlanÄ±yor...")
drive.mount('/content/drive')

# Veri yollarÄ±
BASE_PATH = '/content/drive/MyDrive/archive'
TRAIN_PATH = os.path.join(BASE_PATH, 'train')
TEST_PATH = os.path.join(BASE_PATH, 'test')

# Hiperparametreler (DENGE: HÄ±z + Ä°yi SonuÃ§lar)
IMG_SIZE = 48
BATCH_SIZE = 64
EPOCHS = 50
LEARNING_RATE = 0.0005  # Daha stabil Ã¶ÄŸrenme

print(f"\n[2/9] Hiperparametreler ayarlandÄ±:")
print(f"  - GÃ¶rÃ¼ntÃ¼ Boyutu: {IMG_SIZE}x{IMG_SIZE}")
print(f"  - Batch Boyutu: {BATCH_SIZE}")
print(f"  - Epoch SayÄ±sÄ±: {EPOCHS}")
print(f"  - Ã–ÄŸrenme HÄ±zÄ±: {LEARNING_RATE}")

# Veri yÃ¼kleme fonksiyonu - GELÄ°ÅTÄ°RÄ°LMÄ°Å
def load_data_from_folder(folder_path, target_samples_per_class=1000):
    """Bir klasÃ¶rden verileri yÃ¼kle - histogram eÅŸitleme ile"""
    images = []
    labels = []
    emotion_labels = []

    print(f"\nKlasÃ¶r yÃ¼kleniyor: {folder_path}")

    for emotion_folder in sorted(os.listdir(folder_path)):
        emotion_path = os.path.join(folder_path, emotion_folder)

        if os.path.isdir(emotion_path):
            if emotion_folder not in emotion_labels:
                emotion_labels.append(emotion_folder)

            label_idx = emotion_labels.index(emotion_folder)

            image_files = [f for f in os.listdir(emotion_path)
                          if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

            image_files = image_files[:target_samples_per_class]

            loaded_count = 0
            for img_file in image_files:
                img_path = os.path.join(emotion_path, img_file)
                try:
                    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                    if img is not None:
                        # Histogram eÅŸitleme - daha iyi Ã¶zellik Ã§Ä±karÄ±mÄ±
                        img = cv2.equalizeHist(img)
                        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
                        images.append(img)
                        labels.append(label_idx)
                        loaded_count += 1
                except:
                    continue

            print(f"  âœ“ {emotion_folder}: {loaded_count} gÃ¶rÃ¼ntÃ¼ yÃ¼klendi")

    images = np.array(images).reshape(-1, IMG_SIZE, IMG_SIZE, 1) / 255.0
    labels = np.array(labels)

    return images, labels, emotion_labels

# EÄŸitim ve test verilerini yÃ¼kle
print(f"\n[3/9] Veriler yÃ¼kleniyor...")
print("=" * 60)

# Train verisi - yeterli veri
X_train, y_train, emotion_names = load_data_from_folder(TRAIN_PATH, target_samples_per_class=1000)

# Test verisi
X_test, y_test, _ = load_data_from_folder(TEST_PATH, target_samples_per_class=200)

print(f"\nğŸ“Š VERÄ° SETÄ° YÃœKLEME SONUÃ‡LARI:")
print(f"  - EÄŸitim Seti: {len(X_train)} gÃ¶rÃ¼ntÃ¼")
print(f"  - Test Seti: {len(X_test)} gÃ¶rÃ¼ntÃ¼")
print(f"  - Toplam: {len(X_train) + len(X_test)} gÃ¶rÃ¼ntÃ¼")
print(f"  - Duygu Kategorileri: {emotion_names}")

# Data Augmentation - Dengeli
train_datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.15,
    height_shift_range=0.15,
    horizontal_flip=True,
    zoom_range=0.15,
    shear_range=0.15,
    fill_mode='nearest'
)

# CNN Modeli - GÃœÃ‡LENDÄ°RÄ°LMÄ°Å MÄ°MARÄ°
print(f"\n[4/9] GeliÅŸmiÅŸ model oluÅŸturuluyor...")

model = models.Sequential([
    # Blok 1
    layers.Conv2D(64, (3, 3), padding='same', input_shape=(IMG_SIZE, IMG_SIZE, 1)),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Conv2D(64, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.25),

    # Blok 2
    layers.Conv2D(128, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Conv2D(128, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.25),

    # Blok 3
    layers.Conv2D(256, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Conv2D(256, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.25),

    # Blok 4
    layers.Conv2D(512, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Conv2D(512, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.25),

    # Dense katmanlar
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(256, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.5),
    layers.Dense(len(emotion_names), activation='softmax')
])

# Model derleme
model.compile(
    optimizer=Adam(learning_rate=LEARNING_RATE),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("\n  Model Mimarisi:")
model.summary()

# Callbacks - optimize
callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        patience=10,
        restore_best_weights=True,
        verbose=1,
        mode='max'
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=5,
        min_lr=1e-7,
        verbose=1
    ),
    ModelCheckpoint(
        'best_model.h5',
        monitor='val_accuracy',
        save_best_only=True,
        verbose=1,
        mode='max'
    )
]

# Model eÄŸitimi
print(f"\n[5/9] Model eÄŸitimi baÅŸlÄ±yor...")
print("=" * 60)

start_time = time.time()

history = model.fit(
    train_datagen.flow(X_train, y_train, batch_size=BATCH_SIZE),
    steps_per_epoch=len(X_train) // BATCH_SIZE,
    validation_data=(X_test, y_test),
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1
)

training_time = time.time() - start_time
print(f"\nâœ“ EÄŸitim tamamlandÄ±! SÃ¼re: {training_time/60:.2f} dakika")

# Model deÄŸerlendirme
print(f"\n[6/9] Model test ediliyor...")
print("=" * 60)

test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=0)
y_pred = np.argmax(model.predict(X_test, verbose=0), axis=1)

print(f"\nğŸ“Š TEST SONUÃ‡LARI:")
print(f"  â€¢ Test KaybÄ±: {test_loss:.4f}")
print(f"  â€¢ Test DoÄŸruluÄŸu: {test_accuracy*100:.2f}%")

# Classification Report
print(f"\n[7/9] DetaylÄ± Performans Metrikleri:")
print("=" * 60)
print("\nSINIFLANDIRMA RAPORU:")
print(classification_report(y_test, y_pred, target_names=emotion_names, digits=4))

# Confusion Matrix
print(f"\n[8/9] Confusion Matrix oluÅŸturuluyor...")
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=emotion_names,
            yticklabels=emotion_names,
            cbar_kws={'label': 'GÃ¶rÃ¼ntÃ¼ SayÄ±sÄ±'})
plt.title('Confusion Matrix - KarÄ±ÅŸÄ±klÄ±k Matrisi', fontsize=16, fontweight='bold')
plt.ylabel('GerÃ§ek Duygu (Actual)', fontsize=12)
plt.xlabel('Tahmin Edilen Duygu (Predicted)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# EÄŸitim grafiklerini Ã§iz
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# DoÄŸruluk grafiÄŸi
axes[0].plot(history.history['accuracy'], label='EÄŸitim DoÄŸruluÄŸu', linewidth=2)
axes[0].plot(history.history['val_accuracy'], label='Validasyon DoÄŸruluÄŸu', linewidth=2)
axes[0].set_title('Model DoÄŸruluÄŸu', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('DoÄŸruluk (Accuracy)')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# KayÄ±p grafiÄŸi
axes[1].plot(history.history['loss'], label='EÄŸitim KaybÄ±', linewidth=2)
axes[1].plot(history.history['val_loss'], label='Validasyon KaybÄ±', linewidth=2)
axes[1].set_title('Model KaybÄ±', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('KayÄ±p (Loss)')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('training_history.png', dpi=300, bbox_inches='tight')
plt.show()

# SÄ±nÄ±f bazÄ±nda doÄŸruluk
print(f"\n[9/9] SÄ±nÄ±f bazÄ±nda performans hesaplanÄ±yor...")
print("\nğŸ“ˆ SINIF BAZINDA DOÄRULUK:")
print("=" * 60)
for i, emotion in enumerate(emotion_names):
    mask = y_test == i
    if np.sum(mask) > 0:
        class_accuracy = np.sum((y_test == i) & (y_pred == i)) / np.sum(mask)
        class_total = np.sum(mask)
        print(f"  {emotion:20s}: {class_accuracy*100:6.2f}% ({class_total} test gÃ¶rÃ¼ntÃ¼sÃ¼)")

# Ã–zet istatistikler
print("\n" + "=" * 60)
print("ğŸ“‹ Ã–ZET Ä°STATÄ°STÄ°KLER")
print("=" * 60)
print(f"Toplam Veri: {len(X_train) + len(X_test)} gÃ¶rÃ¼ntÃ¼")
print(f"EÄŸitim Verisi: {len(X_train)} gÃ¶rÃ¼ntÃ¼")
print(f"Test Verisi: {len(X_test)} gÃ¶rÃ¼ntÃ¼")
print(f"Duygu SayÄ±sÄ±: {len(emotion_names)}")
print(f"EÄŸitim SÃ¼resi: {training_time/60:.2f} dakika")
print(f"En Ä°yi Validasyon DoÄŸruluÄŸu: {max(history.history['val_accuracy'])*100:.2f}%")
print(f"Final Test DoÄŸruluÄŸu: {test_accuracy*100:.2f}%")
print(f"Model Parametreleri: {model.count_params():,}")
print("=" * 60)

# Model kaydetme
model.save('yuz_ifadesi_model_final.h5')
print("\nâœ… Model 'yuz_ifadesi_model_final.h5' olarak kaydedildi!")
print("âœ… Confusion Matrix 'confusion_matrix.png' olarak kaydedildi!")
print("âœ… EÄŸitim grafikleri 'training_history.png' olarak kaydedildi!")
print("\nğŸ‰ Ä°ÅŸlem baÅŸarÄ±yla tamamlandÄ±!")
```
